```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
df = pd.DataFrame(iris_data)
df.drop('Species', inplace=True, axis=1)
```

```python
plt.figure(figsize=(6,6))
```
